# 5. Assess the trade-off between risk and return by plotting risk metrics against historical returns.

```python
import numpy as np
```

```python
aapl['date'] = pd.to_datetime(aapl['date'])


aapl['Returns'] = aapl['close'].pct_change() * 100  

risk_metric = aapl['Returns'].std()

plt.figure(figsize=(16, 10))
plt.scatter([risk_metric] * len(aapl), aapl['Returns'], color='r', label='Data Points')
plt.xlabel('Risk (Standard Deviation of Returns)')
plt.ylabel('Historical Returns (%)')
plt.title('Risk-Return Analysis')
plt.grid(True)


plt.annotate(f'Risk Metric: {risk_metric:.2f}', (risk_metric, aapl['Returns'].max()), textcoords="offset points", xytext=(0, 10), ha='center')

plt.legend()
plt.show()
```

# 6. Compare the stock's performance to relevant market indices to assess its relative strength

### Comparison 2013 to 2018

```python
googl = pd.read_csv('/Users/eltonmungai/Documents/Portfolio/datasets/archive/individual_stocks_5yr/individual_stocks_5yr/GOOGL_data.csv')
googl
```

```python
aapl['date'] = pd.to_datetime(aapl['date'])
googl['date'] = pd.to_datetime(googl['date'])

aapl['Apple Returns'] = aapl['close'].pct_change() * 100 
googl['Google Returns'] =googl['close'].pct_change() * 100

merged_data = pd.merge(aapl[['date', 'Apple Returns']], googl[['date', 'Google Returns']], on='date', how='inner')
merged_data['Excess Returns'] = merged_data['Apple Returns'] - merged_data['Google Returns']
```

```python
merged_data
```

```python
plt.figure(figsize=(10, 6))
plt.plot(merged_data['date'], merged_data['Apple Returns'], label='Apple Returns', color='blue')
plt.plot(merged_data['date'], merged_data['Google Returns'], label='Google Returns', color='red')
plt.xlabel('Date')
plt.ylabel('Daily Returns (%)')
plt.title('Apple vs. Google  Performance')
plt.legend()
plt.grid(True)
plt.show()
```

## Comparison in 2013

```python
aapl['date'] = pd.to_datetime(aapl['date'],  format='%d/%m/%Y')
googl['date'] = pd.to_datetime(aapl['date'],  format='%d/%m/%Y')

start_date = '01/01/2013'
end_date = '31/12/2013'

start_date = pd.to_datetime(start_date, format='%d/%m/%Y')
end_date = pd.to_datetime(end_date, format='%d/%m/%Y')

aapl_2013 = aapl[(aapl['date'] >= start_date) & (aapl['date'] <= end_date)]
googl_2013 = googl[(aapl['date'] >= start_date) & (googl['date'] <= end_date)]
```

```python
aapl_2013.loc[:,'Apple Returns 2013'] = aapl_2013['close'].pct_change() * 100
googl_2013.loc[:,'Google Returns 2013'] = googl_2013['close'].pct_change() * 100
```

```python
merged_data = pd.merge(aapl_2013[['date', 'Apple Returns 2013']], googl_2013[['date', 'Google Returns 2013']], on='date', how='inner')
merged_data['Excess Returns'] = merged_data['Apple Returns 2013'] - merged_data['Google Returns 2013']

plt.figure(figsize=(10, 6))
plt.plot(merged_data['date'], merged_data['Apple Returns 2013'], label='Apple Returns 2013', color='blue')
plt.plot(merged_data['date'], merged_data['Google Returns 2013'], label='Google Returns 2013', color='red')
plt.xlabel('Date')
plt.ylabel('Daily Returns (%)')
plt.title('Apple vs. Google Index Performance (Jan 1, 2013 - Dec 31, 2013)')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.show()
```
